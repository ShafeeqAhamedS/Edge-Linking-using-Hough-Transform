```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Actual Road Image

```python
image=cv2.imread("roads.png",1)

plt.imshow(image)
plt.title('Original')
plt.axis('off')
```

```python
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

plt.imshow(image,cmap="gray")
plt.title('gray')
plt.axis('off')
```

```python
img = cv2.GaussianBlur(src = gray, ksize = (15,15), sigmaX=0,sigmaY=0)

plt.imshow(img)
plt.title('EDGES')
plt.axis('off')
```

```python
edges = cv2.Canny(img, 120, 150)

plt.imshow(edges)
plt.title('EDGES')
plt.axis('off')
```

```python
lines=cv2.HoughLinesP(edges,1,np.pi/180,threshold=80,minLineLength=50,maxLineGap=250)
```

```python
for line in lines:
    x1,y1,x2,y2=line[0]
    cv2.line(img,(x1,y1),(x2,y2),(0,0,205),2)
```

```python
plt.imshow(img)
plt.title('HOUGH')
plt.axis('off')
```
